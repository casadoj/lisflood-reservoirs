```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from pathlib import Path
```

```python
df = pd.DataFrame({'EFAS5': {'count': 1423, 'storage': 441}, 'EFAS6': {'count': 1594, 'storage': 1056}})

df
```

```python
fig, ax = plt.subplots(figsize=(4, 2))
w = .7
x = np.array([1, 3])
ax.bar(x - w / 2, df.EFAS5, width=w, color='#a6cee3', label='v5')
ax.bar(x + w / 2, df.EFAS6, width=w, color='#f39728', label='v6')
ax.set_xticks(x)
ax.set_xticklabels(['count/n(-)', 'total storage/n(10$^3$ hm³)'])
ax.tick_params(axis='x', length=0)
yticks = np.linspace(0, 2000, 5)
ax.set_yticks(yticks)
ax.spines[['bottom', 'right', 'top']].set_visible(False)
ax.legend(frameon=False);
```
